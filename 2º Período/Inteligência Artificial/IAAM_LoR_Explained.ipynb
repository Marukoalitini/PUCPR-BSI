```python
# %% Let's begin

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Some Synthetic Data
x0 = np.array([-3, -2.8, -2.4, -2, -1.7, -1.5, -1, -0.3, 0.1, 0.6])
y0 = np.zeros(10)

x1 = np.array([-1.4, -1.3, -0.9, -0.5, -0.2, 0.0, 0.5, 1.0, 1.5, 2])
y1 = np.ones(10)

X = np.concatenate([x0, x1])
T = np.concatenate([y0, y1])

# Plot Values
plt.figure(1)
plt.plot(x0, np.full_like(x0, 0.5), 'ok')
plt.plot(x1, np.full_like(x1, 0.5), 'sk')
plt.xlabel('Feature')
plt.ylabel('Output')
plt.show()

```

```python

# Base logistic regression
beta0 = 0
beta1 = 1

out = []
for i in np.arange(-5, 5, 0.01):
    out.append(1 / (1 + np.exp(-beta0 - beta1 * i)))
i = np.arange(-5, 5, 0.01)

# Plot values
plt.figure(2)
plt.plot(x0, np.full_like(x0, 0.5), 'ok')
plt.plot(x1, np.full_like(x1, 0.5), 'sk')
plt.plot(i, out, '-k')
plt.grid(True)
plt.xlabel('Feature')
plt.ylabel('Output')
plt.show()
```

```python
beta0 = 0
beta1 = 1
prediction = []

plt.figure(3)
plt.plot(x0, np.full_like(x0, 0.5), 'ok')
plt.plot(x1, np.full_like(x1, 0.5), 'sk')
plt.plot(i, out, '-k')
plt.grid(True)
plt.xlabel('Feature')
plt.ylabel('Output')

for i in range(len(X)):
    pred = 1 / (1 + np.exp(-beta0 - beta1 * X[i]))
    prediction.append(pred)
    plt.quiver(X[i], 0.5, 0, pred - 0.5, color='k', scale_units='xy', angles='xy', scale=1)
    if pred < 0.5:
        plt.plot(X[i], T[i], '*r')
    else:
        plt.plot(X[i], T[i], '*b')

out = []
for i in np.arange(-4, 4, 0.01):
    out.append(1 / (1 + np.exp(-beta0 - beta1 * i)))
i = np.arange(-4, 4, 0.01)

plt.plot(i, out, '-m')
plt.grid(True)
plt.show()
```

```python
# Confusion Matrix
plt.figure(4)
cm = confusion_matrix(T, np.round(prediction))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
