Exercício 4

O vetor X é um vetor que contém os valores de -5 a 4 crescendo de 1 em 1. A função arange cria um vetor com os seguintes parâmetros: (início, fim, intervalo).

```python
a = 4
b = 7
vetorY = funcao1oGrau(a,b,vetorX)
print(vetorY)
```

Exercício 5

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(vetorX,vetorY,label = "Função 1º Grau")
plt.show()
```

a) Ao traçar uma linha entre os pontos, forma uma reta com inclinação para cima, o que indica que a função é uma função de 1º grau com coeficiente angular positivo.

b)

```python
vetorX = np.arange(-5,5,0.5)
vetorY = funcao1oGrau(a,b,vetorX)
fig = plt.figure(figsize=(10,10))
plt.scatter(vetorX,vetorY,label = "Função 1º Grau")
plt.show()
```
